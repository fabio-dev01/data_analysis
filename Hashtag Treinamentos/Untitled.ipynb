```python
#Importar a Biblioteca
import pandas as pd

#Visualizar Base de Dados
vendas_df = pd.read_excel('Vendas-dez.xlsx')
vendas_df.head()
```

```python
#Quantidade de transações por loja (shopping) OK
#Mostra a tabela de acordo com a escolha da Loja (Shopping)
#Faturamento por loja (shopping) OK
#Buscar a média de QUANTIDADE por compra dos Shoppings
#Saber a média da QUANTIDADE por shopping determinado
```

```python
#Quantidade de transações por loja (Shopping)
transacoes = vendas_df['ID Loja'].value_counts()
print(transacoes)
```

```python
#Mostra a tabela de acordo com a escolha da Loja (Shopping)

vendas_rmar_rec = vendas_df.loc[vendas_df['ID Loja'] == 'Rio Mar Recife', ['ID Loja', 'Produto', 'Valor Final']]
print(vendas_rmar_rec)
```

```python
#Faturamento por Loja (Shopping)
faturamento = vendas_df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(faturamento)
```
